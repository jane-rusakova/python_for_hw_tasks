1. Створіть змінні `a`, `b`, `c` та `d`, задайте їм значення відповідно 101, 210, 111, 402, та обчисліть з допомогою Python наступний вираз:  $$ \frac{(a+b) * c}{d} $$

**Важливо**: тут ми маємо обчислити значення саме з використанням змінних, а не просто чисел, як би ми робили у калькуряторі. Змінні потрібні аби за бажання ми могли змінити їх значення та не переписувати код обчислення виразу, бо він буде працювати для різних значень змінних.

```python
a = 101
b = 210
c = 111
d = 402

result = (a + b) * c / d

print(result)
```

11. Виведіть кількість літер `a` у столиці Мадагаскару `Antananarivo`. Важливо врахувати всі літери, як великі, так і маленькі.

```python
capital = "Antananarivo"

# Робимо всі літери одноакового регістру, щоб порахувати 'A' і 'a'
count_a = capital.lower().count("a")

print(count_a)
```

Очікувана відповідь: 4

12. Створіть рядок `my_string` зі значенням: "The quick brown fox jumps over the lazy dog!!!".   
Перевірте, чи починається рядок `my_string` з "The" і чи закінчується він "dog!". Виведіть результати на екран.

```python
my_string = "The quick brown fox jumps over the lazy dog!!!"

starts_with_the = my_string.startswith("The")

ends_with_dog = my_string.endswith("dog!")

print(starts_with_the)
print(ends_with_dog)
```

13. Розділіть рядок `my_string` на список окремих слів, а потім з'єднайте всі слова через нижнє підкреслювання. Виведіть результат на екран.

```python
words = my_string.split()

joined_string = "_".join(words)

print(joined_string)
```

Очікуваний результат: `The_quick_brown_fox_jumps_over_the_lazy_dog!!!`
